```python
from sklearn.model_selection import train_test_split
```

```python
y=dataframe["gogus_agrı_tipi"].values
x=dataframe.drop("gogus_agrı_tipi",axis=1).values
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model1=Sequential()

model1.add(Dense(11,activation="relu"))
model1.add(Dense(11,activation="relu"))
model1.add(Dense(11,activation="relu"))
model1.add(Dense(11,activation="relu"))
model1.add(Dense(11,activation="relu"))
model1.add(Dense(11,activation="relu"))










model1.add(Dense(1))

model1.compile(optimizer="adam",loss="mse")
```

```python
model1.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
loss=model1.history.history["loss"]
#val_data doğrulama datası
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
kayip_verisi= pd.DataFrame(model1.history.history)
```

```python
kayip_verisi.head()
```

```python
kayip_verisi.plot()
```

```python
trainloss=model1.evaluate(x_train,y_train,verbose=0)
testloss=model1.evaluate(x_test,y_test,verbose=0)
```
